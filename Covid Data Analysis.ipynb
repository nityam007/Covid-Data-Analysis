```python
'''
As majority of the values in the Tottest1M,NewDeaths,ActiveCases,NewCases column are missing, no replacement technique can possibly fill the 
NaN values. The reason behind this is the fact that only any replacement policy requires a majority of the values to be filled,
only then their replacement with average, mean, mode, etc makes sense.

In this case, we will have to proceed with dropping of the column.
'''
```

```python
df=df.drop(["Tottest1M","NewDeaths","ActiveCases","NewCases","numbers"],axis=1)
```
